Only use pd.eval(rate) on rate string values that contain the divisor symbol "/".

```python
# for s in rate_strings:
#     if "/" in s:
#         pd.eval(s)
#         print(s)
    
#     else:
#         s = np.nan
#         print(rate)


rate_strings_with_null=rate_strings.apply(lambda rate: pd.eval(rate) if "/" in rate else np.nan)
rate_strings_with_null.unique()
```

## 3.
