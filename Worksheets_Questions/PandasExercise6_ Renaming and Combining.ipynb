## 1.
`name` in beers and `name` in breweries refer to different sets of information.<br> 
Create a copy of each dataframe with with these columns renamed to `beer_name` and `brewery_name`.

```python
# Your code here
beers_renamed = beers.rename(columns={"name" : "beer_name"})
breweries_renamed = breweries.rename(columns={"name" : "brewery_name"})

# Check your answer
print(beers_renamed.columns)
print(breweries_renamed.columns)
```

## 2.
Brewery numbers 10 and 25 produce 100 different beers between them.<br>(Brewery 10 produces 62 beers and Brewery 25 produces 38)

<br>Create a list of the style of beers each brewery manufatctures and combine them into one list 

```python
brewery_10_beer_styles = beers_renamed[beers_renamed.brewery_num==10]['style']
brewery_25_beer_styles = beers_renamed[beers_renamed.brewery_num==25]['style']

combined = pd.concat([brewery_10_beer_styles,brewery_25_beer_styles])
combined

```

Create a `DataFrame` of products styles that both breweries make.

```python
print(len(brewery_10_beer_styles))
print(len(brewery_25_beer_styles))

styles = combined['style']

# Check your answer
styles.value_counts()
```
