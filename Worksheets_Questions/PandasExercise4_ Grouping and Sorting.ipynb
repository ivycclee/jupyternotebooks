## 1.
What club has the most strikers? Create a `Series` whose index is the `Club` from the dataset, and whose values count how many Strikers ("ST") there are.

## 2.
Create a `Series` whose index is Nationality and whose values is the count,min,max and average Overall score the players from that country got.

```python
national_ratings = fifa.groupby('Nationality').Overall.agg(['count', 'min', 'max'])

# Check your answer
national_ratings
```

## 3.
Do countries specialise in producing professional players for specific positions? Create a `Series` whose index is a `MultiIndex`of `{Nationality, Position}` pairs.Sort the values in the `Series` in descending order based on player count.

```python
country_position_counts = fifa.groupby(['Nationality', 'Position']).Nationality.agg('count').sort_values(ascending=False)

# Check your answer
country_position_counts
```
